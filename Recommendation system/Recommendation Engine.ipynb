```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("book.csv")
dataset.head()
```

```python
dataset_1=dataset.drop(columns=['Unnamed: 0'])
```

```python
dataset_1= dataset_1.rename(columns={'User.ID':'User_ID','Book.Title':'Book_Title','Book.Rating':'Book_Rating'})
```

```python
user_book=dataset_1.pivot_table(index='User_ID',
                          columns='Book_Title',
                          values='Book_Rating').reset_index(drop=True)
```

```python
user_book
```

```python
user_book.index=dataset_1.User_ID.unique()
```

```python
user_book
```

```python
#Import those NaNs with 0 values
user_book.fillna(0,inplace=True)
```

```python
user_book
```

```python
import sklearn.metrics 
```

```python
user_sim = 1 - sklearn.metrics.pairwise_distances( user_book.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
```

```python
user_sim_df.index = dataset_1.User_ID.unique()
user_sim_df.columns = dataset_1.User_ID.unique()
```
